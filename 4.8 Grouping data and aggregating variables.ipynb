## Creatie a loyalty flag for existing customers using the transform() and loc() functions. Label as 'Loyal customer', 'New customer' and 'Regular customer' using the criteria from the exercise

```python
ords_prods_merge ['max_order'] = ords_prods_merge.groupby(['user_id']) ['order_number'].transform('max')
```

```python
ords_prods_merge.loc [ords_prods_merge['max_order'] >40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc [ords_prods_merge ['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.loc [(ords_prods_merge ['max_order'] <=40) & (ords_prods_merge ['max_order'] >10), 'loyalty_flag'] = 'Regular customer' 
```

```python
#check output
ords_prods_merge ['loyalty_flag'].value_counts (dropna = False)
```

## The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. ## The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:.”

### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merge['average_price_of_purchase'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc [ords_prods_merge['average_price_of_purchase'] <10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc [ords_prods_merge['average_price_of_purchase'] >=10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge ['spending_flag'].value_counts (dropna = False)
```

## In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:”.”

## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer”

```python
ords_prods_merge['average_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.loc [ords_prods_merge['average_days_between_orders'] >20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc [(ords_prods_merge['average_days_between_orders'] >10)
& (ords_prods_merge ['average_days_between_orders'] <=20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc [ords_prods_merge['average_days_between_orders'] <=10, 'frequency_flag'] = 'Frequent_customer'
```

```python
#checking
ords_prods_merge['average_days_between_orders'].value_counts (dropna = False)
```
